```python
from roboflow import Roboflow
from dotenv import load_dotenv
import os

load_dotenv()

api_key = os.getenv('API_ROBOFLOW')
workspace_name = os.getenv("ROBOFLOW_WORKSPACE_NAME")
project_name = os.getenv("ROBOFLOW_PROJECT_NAME_devastator")
devastator_version =os.getenv("devastator_version")
print(api_key)
print(workspace_name)
print(project_name)
rf = Roboflow(api_key)

project = rf.workspace(workspace_name).project(project_name)


version = project.version(devastator_version)


dataset = version.download("yolov8")

print("Dataset downloaded into:", dataset.location)
```

```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt') 

results = model.train(
    data=project_name + "-" + devastator_version+ "/data.yaml", 
    epochs=50, 
    imgsz=640,
    device=0 
)
```

```python
exported_model = model.export(format="onnx")
```
